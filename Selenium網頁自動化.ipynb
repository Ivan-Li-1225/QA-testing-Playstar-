### Selenium 網頁自動化練習_2 (自動登入/登出帳號_Microsoft)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.get("https://www.microsoft.com/zh-tw")
time.sleep(2)

icon = driver.find_element(By.XPATH, '//*[@id="uhfLogo"]')
time.sleep(2)

icon_1 = driver.find_element(By.ID, 'close-epb')   # 首頁按下 "不, 謝謝"
icon_1.click()

# icon_2 = driver.find_element(By.ID, 'lightbox-cover')

login_icon = driver.find_element(By.XPATH, '//*[@id="mectrl_main_trigger"]')  # 點選登入鍵
login_icon.click()

account = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.ID, "i0116")))
account.send_keys("ahc600")
time.sleep(2)

# account = driver.find_element(By.ID, 'i0116')
# account.send_keys("ahc600")
# time.sleep(2)


next_step = driver.find_element(By.XPATH, '//*[@id="i0281"]/div[4]/div/div/div/div')  # 輸入帳號後點選下一步
next_step.click()

password = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.ID, "i0118")))
password.send_keys("Koe1050259")
time.sleep(2)

login = driver.find_element(By.XPATH, '//*[@id="i0281"]/div[5]/div/div/div/div')  # 輸入密碼後點選登入
login.click()
print("您已登入Microsoft!")

login_keep = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="pageContent"]/div/form/div[3]/div[2]/div/div[1]')))  # 點選不保持登入
login_keep.click()

logout_icon = driver.find_element(By.CLASS_NAME, 'mectrl_topHeader')
logout_icon.click()

logout_click = driver.find_element(By.ID, 'mectrl_body_signOut')
logout_click.click()
print("您已正常登出Microsoft!")


Relogin = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="mectrl_main_trigger"]')))  # 點選登入鍵(確認已正常登出)
Relogin.click()

# login_icon = driver.find_element(By.XPATH, '//*[@id="mectrl_main_trigger"]')  # 再次點選登入鍵(確認已正常登出)
# login_icon.click()

time.sleep(5)
driver.quit()
```
